```python
import json
import random
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def extract_data(path):
  data = []
  with open(path, 'r') as f:
      texts = f.readlines()

  while("" in texts):
      texts.remove("")
  for i, text in enumerate(texts):
      text = text.strip()
      text = text.replace("**", "").strip()
      text = text.replace("-", "").strip()
      if text == "":
          continue
      text = text.split(":")[-1].strip()
      data.append(text)
  return data

d1 = extract_data('/content/drive/MyDrive/Badminton ChatBot/badminton-data.txt')
d2 = extract_data('/content/drive/MyDrive/Badminton ChatBot/badminton-data2.txt')
data = d1 + d2
```

```python
paied_data = []
for x in range(1, len(data),2):
    paied_data.append((data[x-1], data[x]))
random.shuffle(paied_data)
```

```python
split = int(0.85 * len(paied_data))
train = []
val = []
for q,a in paied_data[:split]:
    train.append({'question':q, 'answer':a})
for q,a in paied_data[-split:]:
    val.append({'question':q, 'answer':a})
```

```python
dataset = {'train':train, 'val':val}
with open(r"C:\Users\chakumar\Downloads\my docs\badmintondata.json", 'w') as f:
    json.dump(dataset,f, indent = 4)
```
